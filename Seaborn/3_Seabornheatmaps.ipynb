```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

# **Heatmaps**

```python
flight_df = sns.load_dataset('flights')
flight_df
```

```python
flight_df = flight_df.pivot("month", "year", "passengers")
flight_df.head()
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df)
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot=True, fmt='d')
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot=True, fmt='d', linecolor='k', linewidth='5')
ax
```

```python
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot=True, fmt='d', linecolor='w', linewidth='5', cmap='Reds')
ax
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(flight_df, cbar=False)
```
