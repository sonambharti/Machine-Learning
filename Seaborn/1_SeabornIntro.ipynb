# **Histogram**

Histogram represent the data distribution by forming bins
along the range of the data and then drawing bars to show
the number of observations that the fail in each bin.

```python
seaborn_df1.describe()
```

```python
bins = [2,3,4,5,6,7,8,9,10]
sns.distplot(seaborn_df1['time_spend_company'], bins=bins)
plt.xticks(bins)
```

```python
sns.distplot(seaborn_df1['time_spend_company'],
             bins=bins,
             kde=False)                             # kde - Kernnel Density Estimate
```

```python
sns.distplot(seaborn_df1['time_spend_company'],
             bins=bins,
             hist=False)                             # hist - histogram
```

```python
sns.distplot(seaborn_df1['time_spend_company'],
             bins=bins,
             rug=True)                             # shows more specific data
plt.xticks(bins)
```

```python
sns.distplot(seaborn_df1['time_spend_company'],
             bins=bins,
             rug=True,
             hist_kws={'color':'red', 'edgecolor':'green', 'linewidth':3, 'alpha':0.5}
             )
plt.xticks(bins)
```
