```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

### Reading Image using Matplotlib

```python
img = mpimg.imread('./content/stego.png')
```

```python
plt.imshow(img)
plt.show()
```

### Reading Image using OpenCV

```python
import cv2
```

```python
imgcv2 = cv2.imread('./content/stego.png')

# Check if the image was loaded successfully
if imgcv2 is None:
    raise FileNotFoundError("Image not found or path is incorrect.")
```

```python
plt.imshow(imgcv2)
plt.show()
```

```python
imgcv2 = cv2.cvtColor(imgcv2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgcv2)
plt.show()
```

### Changing Aspect

```python
plt.imshow(img, aspect = 0.5)
plt.show()
```

```python
img1 = mpimg.imread('/content/1(3).jpg')
img2 = mpimg.imread('/content/1(4).jpg')
img3 = mpimg.imread('/content/1(5).jpg')
img4 = mpimg.imread('/content/1(6).jpg')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)

plt.show()
```

### Color Bar
