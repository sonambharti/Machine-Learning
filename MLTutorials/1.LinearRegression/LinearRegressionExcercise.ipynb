```python
# %matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income(US $)')
plt.plot(df, color='red', marker='o')
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
new_df[:10]
```

```python
income = df.iloc[:, 1]
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, income)
```

```python
reg.predict([[1995]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income(US $)')
# plt.plot(df, color='red', marker='o')
plt.plot(df.year, reg.predict(df[['year']]), color='green')
```
