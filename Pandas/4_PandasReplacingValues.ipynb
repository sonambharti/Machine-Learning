```python
import pandas as pd
```

```python
df_sample = pd.read_csv('Dataset/sample.csv')
df_sample
```

```python
df_sample.fillna(method = 'bfill', inplace=True)      # inplace will update the original dataframe and method is backword fill (i.e. fill null value with next row)
df_sample
```

```python
df_sample.replace(to_replace=26, value=30)
```

```python
# we can also directly provide the numbers

df_sample.replace(50, 4000)
```

```python
# We can also replace a list of data with a particular value

df_sample.replace(to_replace=[50,52,53], value='A')
```

```python
# We can also replace a list of data with a list of values

df_sample.replace(to_replace=[50,52,53], value=['A','B','C'])
```

```python
# We can also replace a list of data with a list of values in a particular column

df_sample['Physics'].replace(to_replace=[50,52,53], value=['A','B','C'])
```

```python
# We can also replace a list of regex data with a list of values

df_sample.replace('[A-Za-z]', 0, regex=True)
```
