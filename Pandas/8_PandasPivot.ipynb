```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/sample2.csv')
df
```

# **Analysing data using Pandas**

pivot_table()

```python
pd.pivot_table(df, index='Branch')          # this table will have average / mean value of eac column w.r.t branch
```

```python
# returning Total No. of marks students got in each Branch
pd.pivot_table(df, index='Branch', aggfunc='sum')           # changing aggregate function from default (mean) to sum
```

```python
# returning No. of students in each Branch
pd.pivot_table(df, index='Branch', aggfunc='count')           # changing aggregate function from default (mean) to count
```

```python
pd.pivot_table(df, index='Branch', columns='Section')
```
